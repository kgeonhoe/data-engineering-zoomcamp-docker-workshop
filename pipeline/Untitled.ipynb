```python
import pandas as pd
import polars as pl
```

```python
df = pl.read_parquet("./green_tripdata_2025-11.parquet")
```

```python
df.filter(pl.col('lpep_pickup_datetime').is_between(
    pl.datetime(2025, 11, 1),
    pl.datetime(2025, 12, 1),
    closed='left'
)).filter(pl.col('trip_distance')<=1).count()
```

```python
df
```

```python
df.filter(pl.col('trip_distance')<=100).group_by(pl.col('lpep_pickup_datetime').dt.date()).agg(pl.col('trip_distance').max()).sort('trip_distance')
```

```python
zonedf = pl.read_csv("./taxi_zone_lookup.csv")
```

```python
zonedf
```

```python
df
```

```python
df.filter(pl.col('lpep_pickup_datetime').dt.date() == pl.datetime(2025, 11, 18)).join(zonedf, left_on='PULocationID', right_on='LocationID', how='left').group_by('Zone').agg(
    pl.col('total_amount').sum()
).sort('total_amount', descending=True).head(1)
```

```python
df.filter(
    pl.col('lpep_pickup_datetime').is_between(
        pl.datetime(2025, 11, 1),
        pl.datetime(2025, 12, 1),
        closed='left'
    )
).join(
    zonedf.select(['LocationID', pl.col('Zone').alias('PUZone')]),
    left_on='PULocationID',
    right_on='LocationID',
    how='left'
).filter(
    pl.col('PUZone') == 'East Harlem North'
).join(
    zonedf.select(['LocationID', pl.col('Zone').alias('DOZone')]),
    left_on='DOLocationID',
    right_on='LocationID',
    how='left'
).select(['tip_amount', 'DOZone']).sort('tip_amount', descending=True).head(1)
```
